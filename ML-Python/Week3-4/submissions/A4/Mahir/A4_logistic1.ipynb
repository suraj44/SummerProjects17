This is a non-regularised implementation of logistic regression.

```python
import numpy as np  #Importing required modules and libraries

from math import *
from numpy import linalg

import matplotlib.pyplot as plt
```

```python
train_data = np.genfromtxt('ex2data1.txt',delimiter=',')
y = train_data[:,-1]
print "Target Values:-"
print y
X= train_data[:,0:2]
m = len(y)
X= np.insert(X,0,np.ones(m),axis = 1)
alpha = 0.005
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def calcost(hypothesis,y):
	sum  = (np.dot(y,np.log(hypothesis))) + np.dot((1-y),np.log(1-hypothesis))
	return -sum/m
```

```python
theta = np.array([0,0,0])
prod = np.dot(X,theta.transpose())
hypothesis = sigmoid(prod)
oldcost = calcost(hypothesis,y)
diff = hypothesis - y
```

```python
while True:
    theta = theta - (alpha/m)*(np.dot(diff.transpose(),X))
    prod = np.dot(X,theta.transpose())
    hypothesis = 1/(1 + np.exp(-prod))
    diff = hypothesis - y
    newcost = calcost(hypothesis,y)
    print newcost
    if abs(newcost-oldcost) <= 0.001*oldcost:
        break
    else:
        oldcost = newcost
```

```python
print "Values of theta:- ",theta
prod = np.dot(X,theta.transpose())
predicted = sigmoid(prod)
print "The predicted values are as follows:- "
print predicted
```
